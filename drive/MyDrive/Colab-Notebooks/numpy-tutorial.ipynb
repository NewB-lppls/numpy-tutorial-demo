```python
#import lib
import numpy as np
```

```python
# Khởi tạo numpy

# 1 khởi tạo cơ bản
x = np.array([[5,3],
              [10,2]])
print(f'Khởi tạo mảng numpy: {x}')

# 2 khởi tạo mảng ngẫu nhiên

y = np.random.randn(3, 3)
print(f'Khởi tạo ngẫu nhiên mãng numpy: \n{y}')

# 3 Khởi tạo mảng 0:

z = np.zeros((4,5))
print(f'Khởi tạo mảng 0:\n{z}')
```

```python
# Save file numpy 

# Định dạng npy
np.save('/content/sample_data', y)
# Định dạng txt
np.savetxt('/content/file.txt', x)
```

```python
# Load file numpyy
# Định dạng npy
y1 = np.load('/content/sample_data.npy')
print(f'Load file numpy định dạng npy:\n{y1}')

# Định dạng txt
x1 = np.loadtxt('/content/file.txt')
print(f'Load file numpy định dạng txt:\n{x1}')

```

```python
# Truy cập numpy

x = np.random.randn(4, 5)
print(f'Khởi tạo mảng ngẫu nhiên:\n{x}')

# Truy cập từng phần tử
print(f'Truy cập phần tử hàng 2 cột 3 của mảng: {x[2,3]}\n')

# Truy cập slice
print(f'Truy cập phần tất cả các phần tử cột 1:\n{x[:,1]}\n')

print(f'Truy cập phần tất cả các phần tử hàng 2:\n{x[2,:]}\n')

print(f'Truy cập phần tử của hàng 2-3:\n{x[2:4,:]}\n')
```

```python
# Thay đổi shape của mảng

x = np.random.randn(3,4)
print(f'Khởi tạo mảng ngẫu nhiên:\n{x}\n')

x1 = x.reshape(6,2)
print(f'Reshape mảng thành 2x6:\n{x1}\n')

#Chuyển vị chiều của mãng

x2 = np.transpose(x)
print(f'Ma trận chuyển vị của x:\n{x2}\n')
```

```python
# Các hàm min max sum
x = np.random.randn(5,4)
print(x)

min_x = x.min()
print(f'Giá trị nhỏ nhất của x: {min_x}')

max_x = x.max()
print(f'Giá trị lớn nhất của x: {min_x}')

sum_x = x.sum()
print(f'Tổng của array x: {sum_x}')
```
